범주형 독립변수와 실수 독립변수의 상호작용을 보이고 있다. 흡연 유무가 실수 독립 변수인 애들에게 말이다 

흡연 유무가 age 하고 bmi의 관계에 영향을 미치고 있다 

흡연 유무에 따라서 다른 독립 변수들의 회귀계수 기울기값이 달라진다는 것을 의미 

그렇다면 흡연 유무를 더미 변수로 두고 

Y_hat  = 흡연을 안한다 : 맞으면 0, 틀리면 1 (회귀 모형) + 흡연을 한다 : 맞으면 0, 틀리면 1 (회귀모형)

```python
df['no_smoke'] = df['smoker'] == 0
df['yes_smoke'] = df['smoker'] == 1
```

```python
df['no_smoke']=  df['no_smoke'].map({True : 1,
                    False : 0})

df['yes_smoke']= df['yes_smoke'].map({True : 1,
                    False : 0})

# 풀랭크 방식으로 기준값은 흡연을 안하는 것
```

```python
df = df.drop(['region','children'], axis = 1) #region 같은 경우에는 선형성을 보이지 않기도 하고 중요해보이지 않으니 빼자 
```

```python
df = df[['age', 'sex', 'bmi', 'no_smoke',
       'yes_smoke','charges']]
df
```

```python
pd.options.display.float_format = '{:.5f}'.format
X = np.matrix(df.iloc[:,:-1])
Y = np.array(df.iloc[:,-1]).reshape(-1,1)
```

```python
theta = np.linalg.inv(X.T.dot(X)).dot(X.T.dot(Y))
theta # 3,4 번째가 흡연 유무에 따른 가중치 
```

```python
age_hat = np.array(theta[0] * X.T[0])
sex_hat = np.array(theta[1] * X.T[1])
bmi_hat = np.array(theta[2] * X.T[2])
```

```python
no_smoke_hat = np.array(theta[3] * X.T[3])
yes_smoke_hat = np.array(theta[4] * X.T[4])
```

```python
Y_hat = no_smoke_hat*(age_hat + bmi_hat) + yes_smoke_hat*(age_hat + bmi_hat) + sex_hat
Y_hat = np.array(Y_hat) 
Y_hat
```

```python
plt.plot(Y[:50],c = 'r',linestyle = '--',lw = 0.5)
plt.plot(Y_hat.T[:50], c = 'b',lw = 0.5)
plt.title('파랑색 : 예측값 \n 빨강색 : 실제값')
# 뭔가 느낌은 비슷한데.. 잔차의 폭이 꽤나 크다 
```

```python
plt.scatter(Y,Y_hat, c = Y)
plt.title('실제값과 예측값의 산점도 ')
plt.xlabel('실제값')
plt.ylabel('예측값')

#실제값과 예측값의 산점도를 보면 선형적이라기보단 비선형 회귀 모형이다.

# 이럴 때는 독립변수를 제곱해줌으로서 다중 선형 회귀 분석이 아니라 다항 회귀 분석으로 시행한다.
```

```python
residual = Y - Y_hat.T

standard_residual = (residual - residual.mean()) / residual.var()**0.5
```

```python
plt.scatter(Y_hat,standard_residual, c = Y_hat )
plt.axhline(y = 0, c = 'r', linestyle = '--')
```

* 잔차가 음의 방향으로 증가하는 경우 다중 선형 회귀 모델이 데이터에 적합하지 않음을 나타냅니다.
* 이는 생략된 변수, 잘못된 기능 형식 또는 데이터의 이상값 때문일 수 있습니다. 
* 이 문제를 해결하려면 관련 변수를 추가하거나 변수를 변환하거나 이상값을 제거하여 모델을 수정해야 할 수 있습니다.
* 또 다른 접근 방식은 독립 변수와 종속 변수 간의 기본 관계를 더 잘 포착하는 다른 모델을 사용하는 것입니다.

# 다중 선형 회귀 분석으로 가자 

```python
df['smoker_squared'] = data['smoker'].map({'yes':2,
                                            'no':1})
```

```python
df_2 = df[['age', 'sex', 'bmi', 'no_smoke', 'yes_smoke','smoker_squared','charges']]
```

```python
X = np.matrix(df_2.iloc[:,:-1])
Y = np.array(df_2.iloc[:,-1])
```

```python
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
```

```python
Y_hat = theta.T[0] * X.T[0] + theta.T[1] * X.T[1]+theta.T[2] * X.T[3]+theta.T[4] * X.T[4]+theta.T[5] * X.T[5] 
```

```python
Y_hat = np.array(Y_hat)
```

```python
plt.figure(figsize = [20,5])
plt.plot(Y[:50])
plt.plot(Y_hat.T[:50])
plt.title('다중 선형 회귀 모형의 예측값') #정확성은 늘어났는데
```

```python
plt.scatter(Y,Y_hat, c = Y)
plt.title('실제값과 예측값의 산점도 ')
plt.xlabel('실제값')
plt.ylabel('예측값')
```

```python
residual = Y.T - Y_hat
standard_residual = (residual - residual.mean()) / residual.var()**0.5

plt.scatter(Y_hat,standard_residual, c = Y_hat )
plt.axhline(y = 0, c = 'r', linestyle = '--') #잔차가 뭐 이따구로 생겼냐 얘는 그냥 다중선형회귀분석으로 불가능한듯 ㅅㄱ 
```
